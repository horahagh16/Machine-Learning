## NumPy


```python
# Q11: Convert a and b to numpy arrays and print their dimension and shape
a = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
b = [[7, 8, 9], [4, 5, 6], [1, 2, 3]]

np_a = np.array(a)
np_b = np.array(b)
print('a dimention:', np_a.ndim)
print('b dimention:', np_b.ndim)
print('a shape:', np_a.shape)
print('b shape:' , np_b.shape)
```

```python
# Q12: Print the matrix multiplication and the element-wise multiplication of a and b
matrix = np.matmul(np_a,np_b)
print('matrix multiplication\n',matrix)
matrix = np.multiply(np_a,np_b)
print('element wise\n',matrix)
```

```python
# Q13: Get the tranpose of a and b and print min, max, sum, std of both of them (transposed matrices)
tr_a = np.transpose(np_a)
tr_b = np.transpose(np_b)
print('transpose a\n',tr_a,'\ntranspose b\n',tr_b)
print('min a',tr_a.min())
print('min b',tr_b.min())
print('max a',tr_a.max())
print('max b',tr_b.max())
print('sum a',tr_a.sum())
print('sum b',tr_b.sum())
print('std a',np.std(tr_a))
print('std b',np.std(tr_b))
```

## Matplotlib

```python
# Q15: Plot x,y1 and x,y2 lines in a figure with figsize (7,6)
# Set the color of x,y1 to blue and x,y2 to green
# Set appropriate title and x,y labels for the plot
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]
y2 = [0, -1, -4, -9, -16, -25]

plt.figure(figsize=(7,6), dpi=100)
plt.plot(x,y1, 'b', label='x^2')
plt.plot(x,y2, 'g', label='-x^2')
plt.legend()
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Q15 Graph!')
```
